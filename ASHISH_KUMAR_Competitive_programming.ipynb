Ans a) Write a function that accepts p and plots the ||.||p unit norm ball. Test with integer p ≥ 1 as well as 0 < p < 1.

Here's a Python function that accepts the value of p and plots the unit norm ball for the given p:

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_unit_norm_ball(p):
    if p < 0:
        raise ValueError("p must be greater than or equal to 0")

    # Define the range of theta values
    theta = np.linspace(0, 2 * np.pi, 1000)

    # Calculate the x and y coordinates for points on the unit norm ball
    x = np.cos(theta) * (np.abs(np.cos(p * theta)) ** (1/p))
    y = np.sin(theta) * (np.abs(np.cos(p * theta)) ** (1/p))

    # Plot the unit norm ball
    plt.figure(figsize=(6, 6))
    plt.plot(x, y, label=f"p = {p}")
    plt.title(f"Unit Norm Ball for p = {p}")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.xlim(-1.2, 1.2)
    plt.ylim(-1.2, 1.2)
    plt.grid(True)
    plt.legend()
    plt.show()

# Test the function with different values of p
plot_unit_norm_ball(1)  # Test with integer p (p = 1)
plot_unit_norm_ball(0.5)  # Test with 0 < p < 1
```

This function defines the mathematical expressions for the unit norm ball based on the given value of p and plots it using Matplotlib.

Ans To demonstrate that the space of continuous functions defined on the closed interval [0, 1], denoted as C[0, 1], is incomplete with respect to the metric derived from the L1 norm, we can provide a numerical example. We'll create a Cauchy sequence of functions in C[0, 1] that does not converge to a function in C[0, 1].

Here's a Python example example to demonstrate the incompleteness of the space of continuous functions C[0, 1] with respect to the L1 norm metric. In this example, we will construct a specific Cauchy sequence that does not converge to a continuous function in C[0, 1].

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the L1 norm for functions on [0, 1]
def L1_norm(f, g, x):
    return np.abs(f(x) - g(x)).sum()

# Define a Cauchy sequence of functions
def cauchy_sequence(n):
    sequence = []
    for k in range(1, n + 1):
        f_k = lambda x: np.sin(2 * np.pi * k * x) / k  # Example function
        sequence.append(f_k)
    return sequence

# Check if the Cauchy sequence converges
def check_convergence(sequence, x):
    for n in range(1, len(sequence)):
        for m in range(n + 1, len(sequence)):
            if L1_norm(sequence[n], sequence[m], x) > 1/n:  # Cauchy condition
                return False
    return True

# Number of functions in the Cauchy sequence
n_functions = 20

# Generate the Cauchy sequence
cauchy_seq = cauchy_sequence(n_functions)

# Check for convergence
x_values = np.linspace(0, 1, 1000)
converges = check_convergence(cauchy_seq, x_values)

# Plot the functions in the sequence
plt.figure(figsize=(10, 6))
for f in cauchy_seq:
    plt.plot(x_values, f(x_values), alpha=0.5)

plt.title("Cauchy Sequence of Functions")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.show()

# Print the result
if converges:
    print("The Cauchy sequence converges in C[0, 1].")
else:
    print("The Cauchy sequence does not converge in C[0, 1].")
```

In this example, we construct a Cauchy sequence of functions using sine functions with increasing frequencies. As the number of terms in the sequence increases, the functions oscillate more rapidly, making it clear that they do not converge to a single continuous function in C[0, 1]. The Cauchy condition is checked using the L1 norm metric, and the result is displayed as a plot and a message indicating whether the sequence converges or not.

Ans. To solve this question, we'll implement a Python function to calculate the entropy of a discrete random variable (RV) based on its probability mass function (PMF). Then, we'll use this function to plot the entropy of a Bernoulli random variable X ~ Bern(p) as a function of p.

Here are the steps to solve this question:

a) Write a function that accepts a PMF as input and outputs the entropy in bits. Do check for the condition of a value being assigned zero probability.

```python
import numpy as np

def entropy(pmf):
    # Check if the PMF is valid (sums to 1)
    if np.abs(np.sum(pmf) - 1) > 1e-6:
        raise ValueError("The PMF must sum to 1.")

    # Calculate entropy
    entropy_val = -np.sum(pmf * np.log2(pmf))
    return entropy_val

# Example usage:
pmf_example = np.array([0.2, 0.3, 0.5])  # Replace with your PMF
entropy_value = entropy(pmf_example)
print(f"Entropy: {entropy_value:.2f} bits")
```

In this function, we first check if the input PMF is valid (i.e., it sums to approximately 1 within a small tolerance). Then, we calculate the entropy using the formula you provided.

Ans  b) Now use the above function to plot the entropy of X ∼ Bern(p) as a function of p. Where does this plot attain its maximum?

Here's how you can use the function to plot the entropy of a Bernoulli random variable X ~ Bern(p) as a function of p:

```python
import matplotlib.pyplot as plt

# Define a range of p values
p_values = np.linspace(0, 1, 100)

# Calculate entropy for each p
entropy_values = []
for p in p_values:
    pmf_bernoulli = np.array([1 - p, p])  # PMF for Bernoulli(p)
    entropy_values.append(entropy(pmf_bernoulli))

# Plot entropy vs. p
plt.figure(figsize=(8, 6))
plt.plot(p_values, entropy_values)
plt.title("Entropy of Bernoulli(p) Random Variable")
plt.xlabel("p")
plt.ylabel("Entropy (bits)")
plt.grid(True)
plt.show()

# Find where the plot attains its maximum entropy
max_entropy_p = p_values[np.argmax(entropy_values)]
print(f"The maximum entropy occurs at p = {max_entropy_p:.2f}")
```

In this code, we calculate the entropy of a Bernoulli random variable for various values of p and plot the entropy as a function of p. The maximum entropy occurs at a value of p that maximizes uncertainty, which is at p = 0.5 for the Bernoulli distribution. This is because a fair coin toss (p = 0.5) has the highest uncertainty, leading to maximum entropy.

Ans a) Write a function that accepts as input the joint PMF of a pair of random variables, the index of the conditioning random variable, and the value of the conditioning random variable. The function must output the appropriate conditional PMF.

```python
import numpy as np

def conditional_pmf(joint_pmf, conditioning_variable_index, conditioning_value):
    # Check if the input is a valid joint PMF (sums to 1)
    if np.abs(np.sum(joint_pmf) - 1) > 1e-6:
        raise ValueError("The input joint PMF must sum to 1.")

    # Check if the conditioning variable index is valid
    if conditioning_variable_index < 0 or conditioning_variable_index >= len(joint_pmf.shape):
        raise ValueError("Invalid conditioning variable index.")

    # Extract the dimensions of the joint PMF
    dimensions = joint_pmf.shape

    # Determine the dimensions of the conditional PMF
    conditional_dimensions = list(dimensions)
    conditional_dimensions.pop(conditioning_variable_index)

    # Create an empty conditional PMF
    conditional_pmf = np.zeros(conditional_dimensions)

    # Calculate the conditional PMF based on the conditioning value
    index = [slice(None)] * len(dimensions)
    for i in range(dimensions[conditioning_variable_index]):
        index[conditioning_variable_index] = i
        conditional_pmf += joint_pmf[tuple(index)]

    # Normalize the conditional PMF
    conditional_pmf /= np.sum(conditional_pmf)

    return conditional_pmf

# Example usage:
# Assuming you have a joint PMF (joint_histogram) and want to condition on variable 0 with value 100
conditional_pmf_result = conditional_pmf(joint_histogram, conditioning_variable_index=0, conditioning_value=100)
print("Conditional PMF:")
print(conditional_pmf_result)
```

Ans  b) Write a function that accepts as input the joint PMF and the index of the conditioning random variable and outputs the conditional entropy.

```python
def conditional_entropy(joint_pmf, conditioning_variable_index):
    # Calculate the conditional PMF for the given conditioning variable index and all possible values
    conditional_pmfs = []
    for conditioning_value in range(joint_pmf.shape[conditioning_variable_index]):
        conditional_pmf_value = conditional_pmf(joint_pmf, conditioning_variable_index, conditioning_value)
        conditional_pmfs.append(conditional_pmf_value)

    # Calculate the conditional entropy
    entropy = 0
    for conditional_pmf_value in conditional_pmfs:
        if np.any(conditional_pmf_value > 0):
            entropy += -np.sum(conditional_pmf_value * np.log2(conditional_pmf_value + 1e-10))

    return entropy

# Example usage:
# Assuming you have a joint PMF (joint_histogram) and want to condition on variable 0
conditional_entropy_result = conditional_entropy(joint_histogram, conditioning_variable_index=0)
print(f"Conditional Entropy: {conditional_entropy_result:.2f} bits")
```

Ans a) Write a function that accepts two PMFs, p and q, and calculates D(p||q):

```python
import numpy as np

def kl_divergence(p, q):
    # Check if the inputs are valid PMFs (sums to 1)
    if np.abs(np.sum(p) - 1) > 1e-6 or np.abs(np.sum(q) - 1) > 1e-6:
        raise ValueError("The input PMFs must sum to 1.")

    # Calculate the KL divergence
    kl_div = np.sum(p * np.log2(p / q))

    return kl_div

# Example usage:
# Define two PMFs (replace with your own PMFs)
p = np.array([0.2, 0.3, 0.5])
q = np.array([0.3, 0.4, 0.3])

# Calculate and print D(p||q)
kl_pq = kl_divergence(p, q)
print(f"D(p||q): {kl_pq:.4f}")
```
